```python
import matplotlib.pyplot as plt
import numpy as np
DPI_SET = 100
```

```python
#效价二分类
X = [0,100,200,300,400,500,600,700,800,900]
Y_EEG = [68.5,67.2,63.0,67.2,64.8,67.0,66.9,69.8,66.1,68.0]
Y_GSR = [67.4,64.6,64.6,66.7,65.1,64.3,64.6,69.0,65.6,66.4]
Y_RSP = [67.4,65.9,62.0,66.4,63.5,65.9,65.1,67.7,64.3,66.4]
Y_PPG = [67.1,65.1,62.8,66.4,65.1,66.7,64.8,69.5,64.3,66.7]
plt.figure(dpi = DPI_SET)
plt.plot(X,Y_EEG,'ro',label="EEG")
plt.plot(X,Y_GSR,'bs',label="GSR")
plt.plot(X,Y_RSP,'g^',label="RSP")
plt.plot(X,Y_PPG,'k+',label="PPG")
plt.grid(False)
plt.xlabel(u"最佳MAE表现模型排名")
plt.ylabel("MAE")
plt.title(u'愉悦度预测最佳十次模型对应的MAE')
plt.legend()
plt.show()
```
